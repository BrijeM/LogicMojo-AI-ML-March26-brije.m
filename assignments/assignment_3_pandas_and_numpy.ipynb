#### Q1. Understanding Feature Distributions (Iris Dataset)

Context: Before training any ML model, we analyze feature distributions.

Dataset: Load Iris dataset from sklearn

Tasks:

1. Extract feature matrix as NumPy array
2. Compute: Mean Median Standard deviation Variance (for each feature)
3. Identify: Which feature has highest variability and why it matters in ML
4. Convert any one feature into shape (n,1) and explain why ML models expect this format

```python
from sklearn.datasets import load_iris
import numpy as np
```

```python
iris = load_iris() #loads data as dict
```

```python
# Extract feature matrix as NumPy array
feature_matrix = iris.data # extracting feature
print("Feature Matrix Shape :" ,feature_matrix.shape)
print("Feature Matrix row :" ,feature_matrix[0])
```

```python
# Extract feature names - remove
feature_names = iris.feature_names
print("Feature Names:",feature_names)
```

```python
# Mean Median Standard deviation Variance (for each feature)
mean = np.mean(feature_matrix,axis=0)
median = np.median(feature_matrix,axis=0)
std=np.std(feature_matrix,axis=0)
variance=np.argmax(np.var(feature_matrix,axis=0))
print("Mean:",mean)
print("Median:",median)
print("SD:",std)
print("Variance:",variance)

print("Max Variance:",np.argmax(variance))
```

##### Identify: Which feature has highest variability and why it matters in ML

1. Petal_length has the highest variance in the feature matrix.
2. It means data is more spread, not constant and so we have more information.
3. When variance is more that means the model can learn patterns better.

```python
#Convert any one feature into shape (n,1) and explain why ML models expect this format

petal_width = feature_matrix[:,3]
print(petal_width.shape)

petal_width_reshape = petal_width.reshape(-1,1)
print(petal_width_reshape.shape)
```

```python
# (150,) → just a list
# (150,1) → proper column matrix
# Most models expect data in 2D format
```

#### Q4. Dataset Understanding (Titanic Dataset)

Context: First step in ML pipeline is dataset inspection.

Dataset: Titanic (Seaborn)

Tasks:

1. Display: head(), tail() info(), describe()
2. Identify: Missing values per column Numerical vs categorical features
3. Explain: Why identifying feature types is important before modeling



```python
import seaborn as sns

# Load dataset
df = sns.load_dataset('titanic')

#print(df.head())
#print(df.tail())
#print(df.info())
#print(df.describe())

# Missing values per column Numerical vs categorical features
import numpy as np

# Numerical features
numerical_cols = df.select_dtypes(include=[np.number]).columns
num_data_missing = df[numerical_cols].isnull().sum()

# Categorical features
categorical_cols = df.select_dtypes(include=['object', 'category', 'bool']).columns
cat_data_missing = df[categorical_cols].isna().sum()

print("Numerical Columns:\n", numerical_cols)
print("\n")
print("Missing values per column for Numerical Columns:\n", num_data_missing)
print("\n")
print(num_data_missing[num_data_missing>0])
print("\n")
print("*****************************************************************************************************")
print("\nCategorical Columns:\n", categorical_cols)
print("\n")
print("Missing values per column for Categorical Columns:\n", cat_data_missing)
print("\n")
print(cat_data_missing[cat_data_missing>0])
print("\n")
```

#### Why identifying feature types is important before modeling
1. Different Data Types Need Different Preprocessing
2. Models Only Understand Numbers - convert categorical → numeric before training
3. Helps Choose the Right Model
4. Prevents Logical Errors
5. Better Feature Engineering

#### Q6. Handling Missing Data (Critical ML Step)

Context: Models cannot handle null values.

Dataset: Titanic

Tasks:

1. Fill missing age using median
2. Fill missing embarked using mode
3. Drop deck column
4. Explain: Why different strategies are used for different columns

```python
df["age"].fillna(df["age"].median(),inplace=True)
df["embarked"].fillna(df["embarked"].mode(),inplace=True)
#df.drop(columns=["deck"],inplace=True)
print(df.head())
```

#### Why Different Strategies Are Used

This is the most important part 👇

1. Numerical Data (age) → Use Median

    Why median?

    Robust to outliers
    Age data can be skewed (few very old passengers)

    Mean can get distorted
    Median gives a more realistic central value

2. Categorical Data (embarked) → Use Mode

    Why mode?

    Categorical data = labels (not numbers)
    You can’t take mean/median of "S", "C", "Q"

    Mode = most frequent category
    Safest assumption for missing values

3. High Missing Data (deck) → Drop Column

    Why drop?

    Too many missing values (~75%+)
    Filling would introduce too much noise

    Better to remove than corrupt the dataset

#### Q13. Custom Feature Transformation using apply()

Dataset: Titanic

Tasks:

1. Create age_group: Child / Adult / Senior
2. Use .apply()
3. Compute survival rate per group

```python
#Create age group
def get_age_group(age):
    if age < 18:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'
```

```python
#use .apply()

titanic['age_group'] = titanic['age'].apply(get_age_group)
```

```python
titanic["age_group"].head()
```

```python
#Compute survival rate per group
survival_by_age_group = titanic.groupby("age_group")["survived"].mean()
print(survival_by_age_group)
```

#### Q14. Error Analysis Simulation (Model Evaluation Thinking)

Context: Understanding prediction errors is key in ML.

Tasks:

1. Generate 1000 random error values using NumPy
2. Compute: Mean Standard deviation
3. Identify outliers: Using mean ± 2*std
4. Explain: What high variance means in model performance

```python
import numpy as np
#Generate 1000 random error values using NumPy
error_values = np.random.randn(1000)
```

```python
mean_error = error_values.mean()
std_error = error_values.std()

print("Mean:",mean_error)
print("Standard Deviation:",std_error)
```

```python
#Identify outliers: Using mean ± 2*std
lower_bound = mean_error - 2 * std_error
upper_bound = mean_error + 2 * std_error

outliers = error_values[(error_values < lower_bound) | (error_values > upper_bound)]

print("Number of outliers:", len(outliers))
```

#### Explain: What high variance means in model performance
Variance in MODEL (Prediction Stability) is Bad

This is about how much the model’s predictions change

Example:
Same input → very different outputs
Model behaves inconsistently

Model is unstable
Often overfitting
Poor generalization

In the above case 56 outliers which is almost 5% and is acceptable
