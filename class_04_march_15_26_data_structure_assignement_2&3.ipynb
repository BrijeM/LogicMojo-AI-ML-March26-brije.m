#### 1.Customer Segmentation Engine
You are given a list of customer dictionaries:

customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]
Task:

- Write a function segment_customer(customer):
- If spend divided by income is greater than 0.6 return "High Value"
- If spend divided by income is between 0.3 and 0.6 return "Medium Value"
- Otherwise return "Low Value"
- Create a final dictionary in the format {customer_id: segment}

```python
customers = [ {"id": 1, "age": 25, "income": 50000, "spend": 20000}, {"id": 2, "age": 45, "income": 120000, "spend": 80000}, {"id": 3, "age": 35, "income": 70000, "spend": 30000} ]
```

```python
def segment_customer(customer):
    value = customer['spend']/customer['income']
    if  value > 0.6:
        return "High Value"
    elif value > 0.3 and value < 0.6:
        return "Medium Value"
    else:
        return "Low Value"

segment_dict = dict()
for customer in customers:
    segment = segment_customer(customer)
    segment_dict.update({customer["id"]:segment})
print(segment_dict)
```

#### 2. Text Cleaning Pipeline
You are given:

texts = [
    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]
Task:

- Write a function clean_text(text):
- Remove special characters
- Remove extra spaces
- Convert text to lowercase
- Apply the function to all elements using loop or map
- Return cleaned list

```python
texts = [ "Hello!!! How are you???", "This is AI class.", "Python@@@is###great" ] 
```

#### 3. Fraud Detection Rule Engine
You are given:

transactions = [
    {"id": 101, "amount": 5000, "location": "India"},
    {"id": 102, "amount": 200000, "location": "Unknown"},
    {"id": 103, "amount": 1000, "location": "India"}
]
Task:

- Write a function is_fraud(transaction)
- If amount is greater than 100000 or location is "Unknown" mark as fraud
- Otherwise mark as safe
- Use lambda with filter or map
- Return only fraud transactions

```python
transactions = [ {"id": 101, "amount": 5000, "location": "India"}, {"id": 102, "amount": 200000, "location": "Unknown"}, {"id": 103, "amount": 1000, "location": "India"} ] 
```

```python
def is_fraud(transaction):
    if transaction["amount"]>100000 or transaction["location"].lower() == "unknown":
        transaction['type'] = "fraud"
        return transaction
    else:
        transaction['type'] = "safe"
    
fraud_transactions = filter(is_fraud,transactions)
print(list(fraud_transactions))
```

#### 4. Feature Engineering

data = [

#### 7. Dictionary-Based Recommendation System
You are given:

user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}
Task:

- Write a function recommend(user)
- Recommend items based on common items between users
- Do not recommend items already purchased by the user
- Use sets and dictionary iteration

```python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}
def recommend(user):
    values = []
    user_items = user_purchases[user]
    
    for k,v in user_purchases.items():
        if k!=user:
            values.extend(v)
    return set(values)-set(user_items)
    
print(recommend("user1"))
```

#### 8. Recursive Decision Tree Simulation
Task:

Write a recursive function decision_tree(score)
- If score is greater than 80 return "Approved"
- If score is between 50 and 80 call the function again with score plus 10
- If score is less than 50 return "Rejected"

```python
def decision_tree(score):
    if score > 80:
        return "Approved"
    elif score >= 50:
        return decision_tree(score + 10)
    else:
        return "Rejected"
```

```python
print(decision_tree(45))
print(decision_tree(60))
```
